```python
import numpy as np
import matplotlib.pyplot as plt
#https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html
```

```python
mu, sigma = 0, 1.5
```

```python
s = np.random.normal(mu, sigma, 2500)
```

```python
plt.figure(figsize=(9, 6))
count,bins,ignored = plt.hist(s, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2*np.pi))*np.exp(-(bins - mu)**2/(2*sigma**2)), linewidth=6, color='r')
plt.grid()
```

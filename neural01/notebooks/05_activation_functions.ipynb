```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def relu(x):
    d = np.copy(x)
    d[d < 0] = 0
    return d

def swish(x, b):
    return x/(1+np.exp(-b*x))

def leaky_relu(x):
    d = np.copy(x)
    d[d < 0] = 0.01*d[d < 0]
    return d
```

```python
t = np.linspace(-5, 5, 100)
```

```python
plt.figure(figsize=(8, 10))
plt.plot(t, relu(t), label="relu")
plt.plot(t, leaky_relu(t), label="leaky relu")
plt.plot(t, swish(t, 2), label="swish")
plt.legend()
```
